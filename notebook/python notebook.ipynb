##### RFM ANALYSIS

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# 1. Database Connection
engine = create_engine(
    "mysql+pymysql://root:root@localhost:3306/consumer360_dw")

query = """
SELECT
    c.customer_id,
    c.region,
    f.order_id,
    t.date_time AS order_date,
    f.total_price
FROM consumer360_dw.fact_sales f
JOIN consumer360_dw.dim_customer_region c
    ON f.customer_key = c.customer_key
JOIN consumer360_dw.dim_time t
    ON f.time_key = t.time_key;"""
```

```python
# 2. Data Extraction (SQL → Pandas)
df = pd.read_sql(query, engine)
df.info()
```

```python
df
```

```python
# 4. RFM Calculation
reference_date = df['order_date'].max()+pd.Timedelta(days=1)
rfm = df.groupby('customer_id').agg({
    'order_date': lambda x: (reference_date - x.max()).days,
    'order_id': 'count',
    'total_price': 'sum'
}).reset_index()
rfm.columns = ['customer_id', 'Recency', 'Frequency', 'Monetary']
```

```python
rfm['Region']=df['region']
```

```python
rfm
```

```python
# 5. RFM Segmentation
rfm['R_score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_score'] = pd.qcut(rfm['Frequency'], 5, labels=[1,2,3,4,5])
rfm['M_score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
```

```python
rfm['RFM_score'] = (
    rfm['R_score'].astype(str) +
    rfm['F_score'].astype(str) +
    rfm['M_score'].astype(str)
)
```

```python
def segment_customer(row):
    if row['R_score'] >= 4 and row['F_score'] >= 4 and row['M_score'] >= 4:
        return 'Champion'
    elif row['R_score'] >= 3 and row['F_score'] >= 3:
        return 'Loyal'
    elif row['R_score'] >= 2:
        return 'Hibernating'
    else:
        return 'Potential'
```

```python
rfm['Segment'] = rfm.apply(segment_customer, axis=1)
```

```python
# rfm.groupby('Segment')['customer_id'].count()
rfm.groupby('Segment')['Monetary'].mean().sort_values(ascending=False)
```

##### MARKET BASKET ANALYSIS

```python
# 6. Market Basket Analysis
query_mba = """
SELECT
    c.customer_id,
    i.item_name
FROM consumer360_dw.fact_sales f
JOIN consumer360_dw.dim_item i
    ON f.item_key = i.item_key
JOIN consumer360_dw.dim_customer c
    ON f.customer_key =c.customer_key;
"""
basket_df = pd.read_sql(query_mba, engine)
```

```python
# to find a top items because the Apriori does 
# not scale well on high-dimensional retail data. 
# To ensure production feasibility, I filtered to top-selling products and increased minimum support
top_items = (
    basket_df['item_name']
    .value_counts()
    .head(30)      # keep top 50 items
    .index
)
```

```python
basket_df_filtered = basket_df[
    basket_df['item_name'].isin(top_items)]
```

```python
basket = (
    basket_df_filtered
    .groupby(['customer_id', 'item_name'])
    .size()
    .unstack(fill_value=0)
)

basket = basket.map(lambda x: 1 if x > 0 else 0)
```

```python
# pip install mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_itemsets = apriori(
    basket,
    min_support=0.03,
    use_colnames=True
)
```

```python
rules = association_rules(
    frequent_itemsets,
    metric="lift",
    min_threshold=1
)
```

```python
strong_rules = rules[
    (rules['confidence'] >= 0.6) &
    (rules['lift'] >= 1.1) &
    (rules['support'] >= 0.02)
].sort_values(by='lift', ascending=False)
```

```python
strong_rules=strong_rules[['antecedents', 'consequents','support', 'confidence', 'lift']]
strong_rules['antecedents'] = strong_rules['antecedents'].astype(str)
strong_rules['consequents'] = strong_rules['consequents'].astype(str)
```

```python
strong_rules.to_sql('market_basket_rules', engine, if_exists='replace', index=False)
```

##### COHORT ANALYSIS

```python
import pandas as pd
engine = create_engine(
    "mysql+pymysql://root:root@localhost:3306/consumer360_dw")

query = """
SELECT 
    c.customer_id,
    t.date_time AS order_date,
    f.total_price
FROM fact_sales f
JOIN dim_customer c ON f.customer_key = c.customer_key
JOIN dim_time t ON f.time_key = t.time_key
"""

df = pd.read_sql(query, engine)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['order_month'] = df['order_date'].dt.to_period('M')
```

```python
df['cohort_month'] = df.groupby('customer_id')['order_month'].transform('min')
```

```python
def cohort_index(df):
    year_diff = df['order_month'].dt.year - df['cohort_month'].dt.year
    month_diff = df['order_month'].dt.month - df['cohort_month'].dt.month
    return year_diff * 12 + month_diff + 1

df['cohort_index'] = cohort_index(df)
```

```python
cohort_data = (
    df.groupby(['cohort_month', 'cohort_index'])
      .agg(active_customers=('customer_id', 'nunique'))
      .reset_index()
)
```

```python
cohort_pivot = cohort_data.pivot_table(
    index='cohort_month',
    columns='cohort_index',
    values='active_customers'
)
```

```python
cohort_size = cohort_pivot.iloc[:, 0]
retention = cohort_pivot.divide(cohort_size, axis=0) * 100
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(30,8))
sns.heatmap(retention, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Customer Retention Cohort Analysis")
plt.ylabel("Cohort Month")
plt.xlabel("Cohort Index")
plt.show()
```

```python
# Save cohort analysis result
EXPORT_PATH ="D:/consumer360_project/powerbi_data/"
cohort_data.to_csv(EXPORT_PATH + "cohort_analysis.csv",index=False)

print("Cohort analysis CSV saved successfully")
```

```python
engine = create_engine(
    "mysql+pymysql://root:root@localhost:3306/consumer360_dw")

query = """
SELECT 
    c.customer_id,
    t.date_time AS order_date,
    f.total_price
FROM fact_sales f
JOIN dim_customer c ON f.customer_key = c.customer_key
JOIN dim_time t ON f.time_key = t.time_key
"""

df = pd.read_sql(query, engine)
```

```python
from lifetimes.utils import summary_data_from_transaction_data

summary = summary_data_from_transaction_data(
    df,
    customer_id_col='customer_id',
    datetime_col='order_date',
    monetary_value_col='total_price',
    observation_period_end=df['order_date'].max()
)
```

```python
from lifetimes import BetaGeoFitter

bgf = BetaGeoFitter(penalizer_coef=0.01)
bgf.fit(summary['frequency'], summary['recency'], summary['T'])
```

```python
summary['predicted_purchases_30days'] = bgf.conditional_expected_number_of_purchases_up_to_time(
    30,
    summary['frequency'],
    summary['recency'],
    summary['T']
)
```

```python
from lifetimes import GammaGammaFitter

ggf = GammaGammaFitter(penalizer_coef=0.01)
ggf.fit(summary['frequency'], summary['monetary_value'])
```

```python
summary['CLV'] = ggf.customer_lifetime_value(
    bgf,
    summary['frequency'],
    summary['recency'],
    summary['T'],
    summary['monetary_value'],
    time=6,
    freq='D',
    discount_rate=0.01
)
```

```python
# Save CLV output
summary.reset_index().to_csv(
    "D:/consumer360_project/powerbi_data/clv_output.csv",
    index=False
)

print("CLV output CSV saved successfully")
```
